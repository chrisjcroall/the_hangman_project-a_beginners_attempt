```python
import random
import string
import time
import pygame

def choose_word():
    """Selects a random word from a local dictionary file"""
    with open("dictionary.txt", "r") as file:
        word_list = file.readlines()
        word = random.choice(word_list).strip()
    return word.lower()

# Hangman drawing stages
hangman_stages = [
    """
       _____
      |     |
            |
            |
            |
            |
    """,
    """
       _____
      |     |
      O     |
            |
            |
            |
    """,
    """
       _____
      |     |
      O     |
      |     |
            |
            |
    """,
    """
       _____
      |     |
      O     |
     /|     |
            |
            |
    """,
    """
       _____
      |     |
      O     |
     /|\\    |
            |
            |
    """,
    """
       _____
      |     |
      O     |
     /|\\    |
     /      |
            |
    """,
    """
       _____
      |     |
      O     |
     /|\\    |
     / \\    |
            |
    """
]

def play_hangman():
    """Plays the hangman game"""
    word = choose_word()
    guessed_letters = []
    tries = 6  # Number of tries allowed

    pygame.mixer.init()  # Initialize pygame mixer
    beep_sound = pygame.mixer.Sound("xxx")  # Load audio file

    print("âš¡âš¡âš¡ WELCOME TO HANGMAN! âš¡âš¡âš¡")
    print("Guess the word:")
    display_word = ["_"] * len(word)
    print(" ".join(display_word))

    while tries > 0:
        guess = input("Enter a letter: ").lower()

        # Check if the input is a single letter
        if len(guess) != 1 or guess not in string.ascii_lowercase:
            print("âŒ Invalid input. Please enter a single letter.")
            continue

        # Check if the letter has already been guessed
        if guess in guessed_letters:
            print("âŒ You already guessed that letter. Try again.")
            continue

        guessed_letters.append(guess)

        if guess in word:
            for i in range(len(word)):
                if word[i] == guess:
                    display_word[i] = guess

            print("âœ… Correct guess!")
            print(" ".join(display_word))

            if "_" not in display_word:
                print("ğŸ‰ğŸ‰ğŸ‰ Congratulations! You guessed the word:", word.upper(), "ğŸ‰ğŸ‰ğŸ‰")
                break
        else:
            tries -= 1
            print("âŒ Wrong guess. You have", tries, "tries left.")
            print(hangman_stages[6 - tries])  # Display hangman drawing

        time.sleep(1)  # Add a delay of 1 second
        beep_sound.play()  # Play audio

    if tries == 0:
        print("ğŸ˜¢ You ran out of tries. The word was:", word.upper(), "ğŸ˜¢")

play_hangman()
```
